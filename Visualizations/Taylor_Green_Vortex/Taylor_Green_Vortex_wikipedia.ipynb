```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
###
a=int(round(2*np.pi,2))
b=int(round(2*np.pi,2))
t=1
nu=1e-6
x=np.linspace(0,a,(a*100)+1)
y=np.linspace(0,b,(b*100)+1)
T=np.linspace(0,t,(a*10)+1)
X,Y=np.meshgrid(x,y)
```

```python
## Velocity compenents
Ft=np.e**(-2*nu*T)
u=np.sin(X)*np.cos(Y)
v=-np.cos(X)*np.sin(Y)

fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot()
ax1.set_title("Taylor-Green Vortex")
plt.streamplot(X,Y,u,v,color=(u**2+v**2)**0.5,density=[5,5],cmap="jet")
plt.show()
plt.savefig("Taylor_Green_Vortex.png")
```

```python
import matplotlib.animation as animation
Ft=np.e**(-2*nu*T)
u=np.sin(X)*np.cos(Y)
v=-np.cos(X)*np.sin(Y)
fig1=plt.figure(figsize=(12,12))
ax=fig1.add_subplot()

tgv=plt.streamplot(X,Y,u,v,color=(u**2+v**2)**0.5,density=[5,5],cmap="jet")
plt.title("Taylor-Green Vortex")
cbar=plt.colorbar((tgv).lines)
cbar.set_label("velocity m/s")
plt.show()
def animate(i):
    u=np.sin(X)*np.cos(Y)*Ft[i]
    v=-np.cos(X)*np.sin(Y)*Ft[i]
    tgv=plt.streamplot(X,Y,u,v,color=(u**2+v**2)**0.5,density=[5,5],cmap="jet")
    return tgv

anim=animation.FuncAnimation(fig1,animate,frames=range(0,len(T)))
writer=animation.PillowWriter(fps=(30))
anim.save("Taylor_Green_Vortex_gif.gif",writer=writer)
```
