```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
###
a=int(round(2*np.pi,2))
b=int(round(2*np.pi,2))
t=10
nu=1e-6
x=np.linspace(-a,a,(a*200)+1)
y=np.linspace(-b,b,(b*200)+1)
T=np.linspace(0,12,(a*200)+1)
X,Y=np.meshgrid(x,y)


plt.figure(figsize=(10,10))
plt.scatter(X,Y)
plt.show()
```

```python
## Velocity compenents
Ft=np.e**(-2*nu*T)
u=np.sin(X)*np.cos(Y)
v=-np.cos(X)*np.sin(Y)

fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot()
ax1.set_title("Taylor-Green Vortex")
plt.streamplot(X,Y,u,v,color=(u**2+v**2)**0.5,density=[5,5],cmap="jet")
plt.show()
plt.savefig("Taylor_Green_Vortex.png")
```

```python
import matplotlib.animation as animation
Ft=np.e**(-2*nu*T)
u=np.sin(X)*np.cos(Y)
v=-np.cos(X)*np.sin(Y)
fig1=plt.figure(figsize=(12,12))
plt.streamplot(X,Y,u,v,color=(u**2+v**2)**0.5,density=[5,5],cmap="jet")
plt.title("Taylor-Green Vortex")
#cbar=plt.colorbar()
def animate(i):
    u=np.sin(X)*np.cos(Y)*Ft[i]
    v=-np.cos(X)*np.sin(Y)*Ft[i]
    plt.streamplot(X,Y,u,v,color=(u**2+v**2)**0.5,density=[5,5],cmap="jet")

anim=animation.FuncAnimation(fig1,animate,frames=range(0,5))
writer=animation.PillowWriter(fps=(30))
anim.save("Taylor_Green_Vortex.gif",writer=writer)
```
