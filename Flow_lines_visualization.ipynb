```python
## Import Modules
import numpy as np
import matplotlib.pyplot as plt

## Dimensions
x=8
y=8
## Mesh Grid
A=np.linspace(-x,x,801)
B=np.linspace(-y,y,801)
X,Y= np.meshgrid(A,B)

## Velocity components
u= X**2 + Y -2
v= X + Y**2 -1
```

```python
## plot Stream lines
fig=plt.figure(figsize=(5,5))
ax1=fig.add_subplot()
vel_plot=plt.streamplot(X,Y,u,v,color=np.sqrt(u**2+v**2),cmap="jet",density=[1,1])
cbar=plt.colorbar((vel_plot).lines)
cbar.set_label("Velocity in m/s")
plt.tight_layout()
plt.show()
```
